## Employement vs Unemployement


```python
from sklearn.preprocessing import normalize

employed = df['Estimated Employed'].iloc[:50]
employed.head(2)
```

```python
unemployed = df['Estimated Unemployment Rate (%)'].iloc[:50]
unemployed.head(2)
```

```python
e = normalize([employed])
print(e)
u = normalize([unemployed])
print(u)
```

```python
e =e.flatten()
print(e)

plt.plot(e)
plt.show()


u = u.flatten()
print(u)

plt.plot(u)
plt.show()
```

```python
plt.plot(e, color='red')
plt.plot(u, color='blue')
plt.xlabel("Employed")
plt.ylabel("Unemployed")
plt.show()
```

```python
# To print the month of unemployment

import datetime as dt
import calendar
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['month_int'] =  df['Date'].dt.month
df['month'] =  df['month_int'].apply(lambda x: calendar.month_abbr[x])
```
